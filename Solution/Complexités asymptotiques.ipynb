```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from ipywidgets import interact
import ipywidgets as widgets

def trace(max):
    x=np.arange(1, max, 0.1)
    
    plt.plot(x, [pow(2, x) for x in x], label="exponentielle")
    plt.plot(x, [pow(x, 3) for x in x], label="cubique")
    plt.plot(x, [pow(x, 2) for x in x], label="quadratique")
    plt.plot(x, [x*np.log2(x) for x in x], label="quasi-linaire")    
    plt.plot(x, x, label="linaire")
    plt.plot(x, [np.log2(x) for x in x], label="logarithmique")
    
    plt.xlabel("$n$", fontsize=16)
    plt.ylabel("$O(n)$", fontsize=16)
    plt.legend()
    plt.xticks(range(1, max))
    plt.gcf().set_size_inches(18, 10)
    plt.show()

interact(trace, max=widgets.IntSlider(min=0, max=50,step=1, value=8, description="$n$ max"));
```
